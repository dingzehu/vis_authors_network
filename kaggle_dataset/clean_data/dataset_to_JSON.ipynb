```python
import pandas as pd
import json
import webcolors
import numpy as np
```

```python
#pd.set_option("display.max_rows", None, "display.max_columns", None)
```

```python
# loading Kaggle dataset
authors = pd.read_csv("nips-19872020/authors.csv")
institutions = pd.read_csv("nips-19872020/institutions.csv")
papers = pd.read_csv("nips-19872020/papers.csv")
```

```python
# select subset of authors
#institutions = institutions.dropna().drop_duplicates(subset=['institutions']).reset_index()
authors_mod = authors.dropna().drop_duplicates(subset=['name_in_database']).drop_duplicates(subset=['institution']).reset_index()
authors_mod = authors_mod[['paper_index', 'name_in_database', 'institution']]
```

```python
authors_mod
```

```python
# replace institutions with color values
for i in range(0, len(authors_mod)):
    color = webcolors.html5_serialize_simple_color(webcolors.html5_parse_legacy_color(authors_mod.institution.values[i]))
    authors_mod.at[i, 'institution_color'] = color
    #institutions['color'] = institutions['color'].str.replace(institutions.color[i],color)
```

```python
authors_mod = authors_mod[['institution','institution_color']]
```

```python
authors = authors.dropna().drop_duplicates(subset=['name_in_database']).drop(columns='name').reset_index()
authors
```

```python
authors_res = authors.merge(authors_mod, how='inner', on='institution')
authors_res
```

```python
result = authors_res.to_json(orient="records")
parsed = json.loads(result)
print(json.dumps(parsed, indent=4, ensure_ascii=False))
```
