```python
import pandas as pd
import numpy as np
from pandas import json_normalize
import json
import time
```

```python
# loading Kaggle dataset
authors = pd.read_csv("authors_list_for_semantic_scholar_id.csv", names=["name"])
#institutions = pd.read_csv("nips-19872020/institutions.csv")
#papers = pd.read_csv("nips-19872020/papers.csv")
```

```python
def getAuthorIdFromSemanticScholar(str):
    import requests

    # Maybe we can somehow design the payload to make the request
    # smaller. I reverse enginiered the following:
    payload = {
        "queryString": str,
        "page": 1,
        "pageSize": 1,
        "sort": "relevance",
        "authors": [],
        "coAuthors": [],
        "venues": [],
    }
    url = "https://www.semanticscholar.org/api/1/search"

    response = requests.post(url, json=payload)
    # More official calls https://api.semanticscholar.org/
    return response.json()["matchedAuthors"]
```

```python
authors_res = authors
```

```python
for i in range(0, 500):
    while True:
        # in case the return is empty
            try:
                if not json_normalize(getAuthorIdFromSemanticScholar(authors_res.iloc[i]['name'])).empty:
                    df = json_normalize(getAuthorIdFromSemanticScholar(authors_res.iloc[i]['name']))
                    authorid = df.iloc[0]['id']
                    # insert found id in the corresponding field
                    authors_res.at[i, 'authorid'] = authorid
            except:
                time.sleep(120)
                continue
            break
```
