```python
import numpy as np
import pandas as pd
import csv
import json
import plotly.graph_objects as go
import networkx as nx
from matplotlib import pyplot as plt
```

```python
# load connection matrix
connection_matrix = pd.read_csv('connection_matrix.csv', header=0, index_col=0)
connection_matrix
```

```python
author_start = 100
author_end = 200
G=nx.from_pandas_adjacency(connection_matrix.iloc[author_start:author_end , author_start:author_end])
```

```python
plt.figure(figsize=(15, 15))
nx.draw_shell(G, with_labels=True)
plt.savefig(f"author_network_from_{author_start}_to_{author_end}.png", dpi=100)
```
